```python
#calculo de similitud entre usuarios
#apply es como sustituir un for, apply es aplicar una funcion por filas o columnas

def cos_similarity(item1, item2):
    item1Values = ~np.isnan(item1)
    item2Values = ~np.isnan(item2)
    allValues = np.logical_and(item1Values,item2Values) # get only existent elements of both vectors
    return np.dot(item1[allValues], item2[allValues])/(np.linalg.norm(item1[item1Values]) * np.linalg.norm(item2[item2Values]))

def pre_cos_similarity(item1, df):
    return df.apply(lambda item2: cos_similarity(item1, item2))

```

```python
Rating.drop('Mean', axis=1, inplace=True) # remove mean column that comes at the end

# replace commas for dots and convert previous string column into float
def processCol(col):
    return col.astype(str).apply(lambda val: val.replace(',','.')).astype(float)
Rating = Rating.apply(processCol)

Rating.head()
```

```python

def cos_similarity(item1, item2):
    item1Values = ~np.isnan(item1)
    item2Values = ~np.isnan(item2)
    allValues = np.logical_and(item1Values,item2Values) # get only existent elements of both vectors
    return np.dot(item1[allValues], item2[allValues])/(np.linalg.norm(item1[item1Values]) * np.linalg.norm(item2[item2Values]))

def pre_cos_similarity(item1, df):
    return df.apply(lambda item2: cos_similarity(item1, item2))

```

```python
df_corr = Rating.apply(lambda item1: pre_cos_similarity(item1, Rating))
df_corr.head()
```

```python
#### Calculos de predicción
def predictRating(userRatings, itemSimilarity):
    userHasRating = ~np.isnan(userRatings)
    return np.dot(userRatings[userHasRating], itemSimilarity[userHasRating])/np.sum(itemSimilarity[userHasRating])

def pre_predictRating(userRatings, df_corr):
    return df_corr.apply(lambda itemSimilarity: predictRating(userRatings, itemSimilarity))

```

```python
predictions = Rating.apply(lambda userRatings: pre_predictRating(userRatings, df_corr), axis=1)
predictions.head()
```

### Distancia de Jaccard

```python
texto1 = 'Córdoba'
```

```python
texto2 = 'corDoBA'
```

```python
token1= [ x.lower() for x in texto1] #.lower transforma texto de mayuscula a minuscula
```

```python
token2= [ x.lower() for x in texto2]
```

```python
print('Las letras del token1 son: {}'.format(token1))
print('Las letras del token2 son: {}'.format(token2))
```

```python
set1 = set(token1) #genero los conjuntos, no tienen un orden dado
```

```python
set2 = set(token2)
```

```python
print('El conjunto 1: {}'.format(set1))
print('El conjunto 2: {}'.format(set2))
```

```python
Inter = set1.intersection(set2)
#Con todos los elementos que están en ambas listas, sin elementos repetidos.
# Es decir, si un elemento está sólo en una lista, no debe aparecer en `c`.
#busco la interseccion entre listas
# Python program to illustrate the intersection  of two lists using set() method 

#otra forma  de hacer intersseccion
#def intersection(lst1, lst2): 
#    return list(set(lst1) & set(lst2)) 
#f_intersec=intersection(a,b)
#print(f_intersec)
```

```python
Inter #calculo la intersección entre los conjuntos de letras. (en caso de ser palabras se busca las palabras comunes)
```

```python
Union = set1.union(set2)
#Con todos los elementos que NO están en ambas listas, sin elementos repetidos. 
#Es decir, si un elemento está en `a` y en `b`, no debe aparecer en `c`.
#def not_intersection(lst1, lst2): 
#    return list(set(lst1) ^  set(lst2)) 
#g_intersec=not_intersection(a,b)
#print(g_intersec)
```

```python
Union
```

```python
Jaccard = len(Inter)/len(Union)
```

```python
print('El coeficiente de Jaccard es {}'.format(Jaccard))
#cuantifico cuanto se parecen un conjunto con otro (ej cordoba con o sin tilde. como se comparan las cosas parecidas)
```

```python
#lo hace para elementos o palabras
def jaccard_unigram( d1, d2):
    t1 = [x.lower() for x in d1.split(' ')]
    t2 = [x.lower() for x in d2.split(' ')]
    
    s1 = set(t1)
    s2 = set(t2)
    return len(s1.intersection(s2))/len(s1.union(s2))
```

```python
#lo puedo hacer en una lista entera. (muchos elementos)
#compara el string general, cada elemento de la lista. osea compara cada elemento de la lista.
#revisar que hace cada linea por separado.
def jaccard_list( d1, d2):
    s1 = set(d1)
    s2 = set(d2)
    return len(s1.intersection(s2))/len(s1.union(s2))
```
