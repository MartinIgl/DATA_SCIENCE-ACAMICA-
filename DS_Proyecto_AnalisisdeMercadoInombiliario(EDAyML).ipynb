Para predecir el precio de las propiedades considero modelos de regresión, debido a que la variable de interés es continua.
Por otra parte las metricas de evaluación deben estar asociadas al análisis de regresión. Es por eso que se elige la metrica Raíz del error cuadrático medio (RMSE):
    
    -Es la métrica más comúnmente utilizada para las tareas de regresión y representa a la raíz cuadrada de la distancia cuadrada promedio entre el valor real y el valor pronosticado. Indica el ajuste absoluto del modelo a los datos, cuán cerca están los puntos de datos observados de los valores predichos del modelo. 
    -Tiene la propiedad útil de estar en las mismas unidades que la variable de respuesta.
    -RMSE es una buena medida de la precisión con que el modelo predice la respuesta, y es el criterio más importante para ajustar si el propósito principal del modelo es la predicción.
    -Los valores más bajos de RMSE indican un mejor ajuste. 
    -Tiene sus limitaciones: El error es una variable aleatoria numérica y no se puede captar todo el comportamiento de una variable aleatoria con una sola agregación de observaciones. El error es solo una variable aleatoria, a menudo es una variable aleatoria muy sesgada. Cuando predecimos resultados sesgados, como precios, ingresos, ventas de artículos y muchos más, lo más probable es que el error también sea sesgado, lo que significa que en la mayoría de los casos el error es muy pequeño, pero hay relativamente pocos ejemplos que pueden tener errores extremadamente grandes. Cuando el error es muy sesgado, el promedio a menudo no dice nada.

```python
from sklearn.model_selection import train_test_split


X=X_prop.copy()

y=(X['price'].values)/float(1000000)  #MI VARIABLE A MODELAR
#Esta división es para interpretar mejor las metricas.(todos los valores están en escala 1:1e6)
X=X.drop("price",axis=1) #MIS ATRIBUTOS CON QUE QUIERO MODELAR EL PRECIO

#genero la separación de muestras de entrenamiento y de testeo en proporción 70-30
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=20)

print(train_x.shape, train_y.shape, test_x.shape, test_y.shape)
```

```python
#modelos de aprendizaje supervisado del tipo regresion
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

# Generar un modelo *benchmark*, en este caso es el modelo de regresión lineal.
linear_model = LinearRegression()
#Entreno un modelo de vecinos más cercanos y un modelo de árbol de decisión con hiperparámetros iniciales a elección.
#Defino mis hiperparametros (max_depth,n_neighbors y n° de atributos (inicialmente utilizo todos los atributos de mi muestra).
tree_regressor = DecisionTreeRegressor(max_depth=3, random_state=20)
knn_regressor = KNeighborsRegressor(n_neighbors=5)

#entreno los modelos con mis datos de entrenamiento
linear_model.fit(train_x, train_y)
tree_regressor.fit(train_x, train_y)
knn_regressor.fit(train_x, train_y)

#########################################################################################
#benchmark model-linear regression
#Predicción
preds_train_y=linear_model.predict(train_x)
preds_test_y=linear_model.predict(test_x)

#analizo el comportamiento del RMSE y el balance entre mis datos de testeo y entrenamiento
metrica_modelo_test=[]
metrica_modelo_train=[]
modelos = ['Regresión lineal', 'Árbol de Decisión', 'Vecinos más cercanos']
for i, model in enumerate([linear_model, tree_regressor, knn_regressor]):
    y_train_pred = model.predict(train_x)
    y_test_pred = model.predict(test_x)
    print(f'Modelo: {modelos[i]}')
    #El cálculo del RMSE para cada conjunto (train y test).
    rmse_train = RMSE(train_y, y_train_pred)
    rmse_test = RMSE(test_y, y_test_pred)
    print(f'Raíz del error cuadrático medio en Train: {rmse_train}')
    print(f'Raíz del error cuadrático medio en Test: {rmse_test}')
    metrica_modelo_test.append(rmse_test)
    metrica_modelo_train.append(rmse_train)


#grafico la metrica en funcion del modelo y los datos train/test
labels = ['Reg. lineal', 'DT', 'KNN']
x = np.arange(len(labels))
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(6,5))

rects1 = ax.bar(x - width/2, metrica_modelo_test, width, label='test')
rects2 = ax.bar(x + width/2, metrica_modelo_train, width, label='train')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('RMSE')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


plt.ylim(0, 0.3)
autolabel(rects1)
autolabel(rects2)
plt.title('Métrica de RMSE en función del Modelo\n Con hiperparámetro a elección')
fig.tight_layout()

plt.show()
#Evalúo la métrica elegida en el conjunto de Test y en el conjunto de Train. 
#Se realiza gráficos de valores reales vs. valores predichos.
#En este caso el KNN muestra las menores metricas de error tanto en test/train. Sin embargo,
#para los valores de entrenamiento se observa que tiende a sobreajustarse y no sería un modelo generalizable.
#Apesar de que la metrica en los datos de prueba es menor, con lo que implica una buena performance de mi modelo.
#Todo lo contrario sucede con la regresión lineal, es decir que según mi benchmark el DT y KNN mejoran.
#Pero en este caso tendería a elegir el DT que es el que está entre los dos modelos para no tender a sobreajustar y poder generalizar más.

#¿Cómo sé si son los parámetros óptimos para cada modelo? Analizo la curva de validación según el RMSE para ver el parámetro óptimo para mejorar el desempeño.
#En la próxima celda se analiza esto y se responderá  ¿cuál modelo utilizaría?¿Por qué?
#En este caso se hace el análisis para los datos de las propiedades juntas. Luego lo veo por cada tipo de propiedades. 
```

Esta desición es así, a pesar que las metricas "empeoraron" su desempeño al realizar la optimización. Sin embargo, el arbol de decisión optimizado genera una disminución del error sustancial y los datos de entrentrenamiento tenderían a sobreajustarse mucho a pesar de que el error de los datos de testeo disminuyen también.
Es decir, que haciendo un balance no elegiría el DTregresor optimizado porque no tendría tanta capacidad de generalización por un posible problema de sobreajuste.

```python

def AnalisisOptimo(X,y):
  """
  Escribo los mismos procesos de antes para hacer el analisis para los principales tipos de propiedades.
  """
  train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=20)
  #calculo los scores para mis parametros optimos
  ScoreRMSE=[]
  #inserto los parametros optimos
  max_depth=15
  n_neighbors=9
  Metricas=get_score_regresion(max_depth,n_neighbors,train_x,test_x,train_y,test_y)  
  ScoreRMSE.append(list(Metricas))
  ScoreRMSE=pd.DataFrame(ScoreRMSE,columns=['Depth','n_neighbors','RMSE_lin_train','RMSE_lin_test','RMSE_DT_train','RMSE_DT_test','RMSE_KNN_M_train','RMSE_KNN_M_test','RMSE_KNN_E_train','RMSE_KNN_E_test'])

  #Grafico la métrica de evaluación en función del modelo optimizado.
  metrica_modelo_test=[round(ScoreRMSE.RMSE_lin_test[0],4),round(min(ScoreRMSE.RMSE_DT_test),4),round(min(ScoreRMSE.RMSE_KNN_M_test),4),round(min(ScoreRMSE.RMSE_KNN_E_test),4)]
  indice_DT_optimo=ScoreRMSE[ScoreRMSE['RMSE_DT_test']== ScoreRMSE['RMSE_DT_test'].min()].index.tolist()
  indice_knn_m_optimo=ScoreRMSE[ScoreRMSE['RMSE_KNN_M_test']== ScoreRMSE['RMSE_KNN_M_test'].min()].index.tolist()
  indice_knn_e_optimo=ScoreRMSE[ScoreRMSE['RMSE_KNN_E_test']== ScoreRMSE['RMSE_KNN_E_test'].min()].index.tolist()
  
  metrica_modelo_train=[round(ScoreRMSE.RMSE_lin_train[0],4),round(ScoreRMSE.RMSE_DT_train[indice_DT_optimo[0]],4),round(ScoreRMSE.RMSE_KNN_M_train[indice_knn_m_optimo[0]],4),round(ScoreRMSE.RMSE_KNN_E_train[indice_knn_e_optimo[0]],4)]
  labels = ['Reg. lineal', 'DT', 'KNN_Manhattan','KNN_Euclides']
  x = np.arange(len(labels))
  width = 0.35  # the width of the bars
  fig, ax = plt.subplots(figsize=(8,5))
  rects1 = ax.bar(x - width/2, metrica_modelo_test, width, label='test')
  rects2 = ax.bar(x + width/2, metrica_modelo_train, width, label='train')
  # Add some text for labels, title and custom x-axis tick labels, etc.
  ax.set_ylabel('RMSE')
  ax.set_xticks(x)
  ax.set_xticklabels(labels)
  ax.legend()
  #Muestro los valores optimos de los hiperparametros
  plt.text(0.65, 0.97, ("Valor óptimo DT Max_depth="+str(max_depth)), ha='right', va='top',transform=ax.transAxes) # position text relative to Axes
  plt.text(0.65, 0.92, ("Valor óptimo KNN (p=1) n_neighbors="+str(n_neighbors)), ha='right', va='top',transform=ax.transAxes) # position text relative to Axes
  plt.text(0.65, 0.87, ("Valor óptimo KNN (p=2) n_neighbors="+str(n_neighbors)), ha='right', va='top',transform=ax.transAxes) # position text relative to Axes
  plt.ylim(0, 0.3)
  def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
  autolabel(rects1)
  autolabel(rects2)
  plt.title('Métrica de RMSE en función del Modelo Con hiperparámetros óptimos')
  fig.tight_layout()
  plt.show()


#Realizo el mísmo análisis para cada tipo de propiedad en particular.
#En este caso utilizaré los mejores hiperparámetros y veo el rendimiento de mis modelos para elegir cuál es el mejor modelo para cada tipo de propiedad.
X=Data_filtrado.copy()
X_dep=X[(X['property_type'].isin(['Departamento']))]
X_dep=pd.DataFrame(X_dep[['bathrooms','surface_covered','price']])
X_dep.dropna(how='any', inplace=True)
y=(X_dep['price'].values)/float(1000000)  #esta división es para interpretar mejor las metricas. Entonces todos los valores se multiplican por 1e6
X_dep=X_dep.drop("price",axis=1)

print('Departamentos')
AnalisisOptimo(X_dep,y)
####### ML Con hiperparámetros óptimos para predecir el precio de PH
X=Data_filtrado.copy()
X_ph=X[(X['property_type'].isin(['PH']))]
X_ph=pd.DataFrame(X_ph[['bathrooms','surface_covered','price']])
X_ph.dropna(how='any', inplace=True)
y=(X_ph['price'].values)/float(1000000)  #esta división es para interpretar mejor las metricas. Entonces todos los valores se multiplican por 1e6
X_ph=X_ph.drop("price",axis=1)

print('PH')
AnalisisOptimo(X_ph,y)


####### ML Con hiperparámetros óptimos para predecir el precio de Casa
X=Data_filtrado.copy()
X_casa=X[(X['property_type'].isin(['Casa']))]
X_casa=pd.DataFrame(X_casa[['bathrooms','surface_covered','price']])
X_casa.dropna(how='any', inplace=True)
y=(X_casa['price'].values)/float(1000000)  #esta división es para interpretar mejor las metricas. Entonces todos los valores se multiplican por 1e6
X_casa=X_casa.drop("price",axis=1)

print('Casas')
AnalisisOptimo(X_casa,y)
```

```python
#Defino librerías y funciones a utilizar en la sección Extra
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, f1_score, precision_score, recall_score


#función utilizada en la sección Extra
def sortedgroupedbar(ax, x,y, groupby, data=None, width=0.0, **kwargs):
    """
    Esta función genera un grafico de barra ordenado, equivalente al sns.barplot pero más flexible.
    """
    order = np.zeros(len(data))
    df = data.copy()
    for xi in np.unique(df[x].values):
        group = data[df[x] == xi]
        a = group[y].values
        b = sorted(np.arange(len(a)),key=lambda x:a[x],reverse=True)
        c = sorted(np.arange(len(a)),key=lambda x:b[x])
        order[data[x] == xi] = c   
    df["order"] = order
    u, df["ind"] = np.unique(df[x].values, return_inverse=True)
    step = width/len(np.unique(df[groupby].values))
    for xi,grp in df.groupby(groupby):
        ax.bar(grp["ind"]-width/2.+grp["order"]*step+step/2.,
               grp[y], label=xi, **kwargs) #width=step
    #ax.legend(title=groupby)
    ax.set_xticks(np.arange(len(u)))
    ax.set_xticklabels(u)
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.set_yscale('log')
    plt.xticks(rotation=45)
    sns.set_palette("RdBu")
```

```python
#calculo los scores para un DT Clasifier
def get_score_DT(max_depth,train_x,test_x,train_y,test_y):
    clf_tree= DecisionTreeClassifier(max_depth=max_depth,random_state=20)
    # Entrenar el modelo    
    clf_tree.fit(train_x,train_y)
    #calculo los scores para los datos test
    # Predecir y evaluar sobre el set de testeo    
    preds_test=clf_tree.predict(test_x)
    ACC_test=accuracy_score(preds_test,test_y)
    f1_test=f1_score(preds_test,test_y,average='weighted')
    ps_test=precision_score(preds_test,test_y,average='weighted')
    rs_test=recall_score(preds_test,test_y,average='weighted')
    #calculo los scores para los datos train
    # Predecir y evaluar sobre el set de entrenamiento
    preds_train=clf_tree.predict(train_x)
    ACC_train=accuracy_score(preds_train,train_y)
    f1_train=f1_score(preds_train,train_y,average='weighted')
    ps_train=precision_score(preds_train,train_y,average='weighted')
    rs_train=recall_score(preds_train,train_y,average='weighted')
    return max_depth,  ACC_test, f1_test, ps_test, rs_test,ACC_train, f1_train, ps_train, rs_train

ScoreDT=[]
#veo los scores para set de datos con DT
for max_depth in range(1,15):
    my_scores_DT=get_score_DT(max_depth,train_x,test_x,train_y,test_y)  
    ScoreDT.append(list(my_scores_DT))
    #puedo imprimirlo por pantalla
    #a,b,c,d,e,f,g,h=get_score_DT(max_depth,train_x,test_x,train_y,test_y)  
    #print('TRAIN Para Max_depth: ', max_depth, 'el ACC es:',a,' el f1:',b,'el ps:',c,'el rc:',d)
    #print('TEST Para Max_depth: ', max_depth, 'el ACC es:',e,' el f1:',f,'el ps:',g,'el rc:',h)


ScoreDT=pd.DataFrame(ScoreDT,columns=['Depth','Acc_test','F1_test','Ps_test','Rs_test','Acc_train','F1_train','Ps_train','Rs_train']) #muestra lista del acc en funcion de la prof del arbol
print(ScoreDT)
ScoreDT.shape
```

```python

####################################################################################
#calculo los scores para un Knn Clasifier
def get_scores_knn(n_neighbors,train_x,test_x,train_y,test_y):
    clf_knn= KNeighborsClassifier(n_neighbors=n_neighbors)
    # Entrenar el modelo 
    clf_knn.fit(train_x,train_y)
    #calculo los scores para los datos test
    # Predecir y evaluar sobre el set de testeo
    preds_test=clf_knn.predict(test_x)
    ACC_test=accuracy_score(preds_test,test_y)
    f1_test=f1_score(preds_test,test_y,average='weighted')
    ps_test=precision_score(preds_test,test_y,average='weighted')
    rs_test=recall_score(preds_test,test_y,average='weighted')
    #calculo los scores para los datos train
    # Predecir y evaluar sobre el set de entrenamiento
    preds_train=clf_knn.predict(train_x)
    ACC_train=accuracy_score(preds_train,train_y)
    f1_train=f1_score(preds_train,train_y,average='weighted')
    ps_train=precision_score(preds_train,train_y,average='weighted')
    rs_train=recall_score(preds_train,train_y,average='weighted')
    return n_neighbors, ACC_test, f1_test, ps_test, rs_test,ACC_train, f1_train, ps_train, rs_train

ScoreKNN=[]
for n_neighbors in range(1,15):
    my_scores_knn=get_scores_knn(n_neighbors,train_x,test_x,train_y,test_y)
    #puedo imprimirlo por pantalla
    #a,b,c,d,e,f,g,h=get_scores_knn(n_neighbors,train_x,test_x,train_y,test_y)  
    #print('TRAIN Para knn: ', n_neighbors, 'el ACC es:',a,' el f1:',b,'el ps:',c,'el rc:',d)
    #print('TEST Para knn: ', n_neighbors, 'el ACC es:',e,' el f1:',f,'el ps:',g,'el rc:',h)
    ScoreKNN.append(list(my_scores_knn))

ScoreKNN#muestra lista del acc en funcion del num de vecinos

ScoreKNN=pd.DataFrame(ScoreKNN,columns=['K-Neighbour','Acc_test','F1_test','Ps_test','Rs_test','Acc_train','F1_train','Ps_train','Rs_train']) #muestra lista del acc en funcion de la prof del arbol
 #muestra lista del acc en funcion de la prof del arbol
ScoreKNN.shape

print(ScoreKNN)

```
